```python
extracted_frames = list(src_folder.glob("*_extracted_frames"))
extracted_frames
```

```python
from video2pdf.utils.directory_manager import DirectoryManager
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
for folder in extracted_frames:
    folder_name = folder.name
    internal_id = "_".join(folder_name.split("_")[:-1])
    video_directory = folder.parent / folder_name.split("_")[0]
    print(video_directory)
    video_path = DirectoryManager.get_video_path(video_directory)
    timestamps = []
    for files in folder.glob("*.jpg"):
        file_name = files.stem
        frame_number = int(file_name.split("_")[1])
        time_stamp = VideoProcessor.get_timestamp_from_frame_number(video_path, frame_number)
        print(time_stamp)
        timestamps.append(time_stamp)

    timestamp_strs = list(map(str, timestamps))
    print(" | ".join(timestamp_strs))
    new_row = {"internal_id": internal_id, "timestamps": timestamps}
    df = pd.concat([df, pd.DataFrame([new_row])])
```

```python
df["internal_id"] = df["internal_id"].apply(lambda x: "_".join(x[:-1]))
```
