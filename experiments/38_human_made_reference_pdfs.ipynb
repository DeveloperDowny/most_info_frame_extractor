```python
from pathlib import Path
import pandas as pd

dir = "/home/vedant/Desktop/glimpsify/most_info_frame_extractor/video2pdf/archives/data_archive_50_python_objects_all_dirs_phash_approval_strategy_v3/"
```

```python
videos_df = pd.DataFrame()
for file in Path(dir).rglob("video_path.txt"):
    video = Path(file.read_text())
    new_row = {
        "video_name": video.stem,
        "video_path": video,
        "cached_processed_frames_path": file.parent / "processed_frames.pkl"
    }
    videos_df = pd.concat([videos_df, pd.DataFrame([new_row])], ignore_index=True)
```

```python
from video2pdf.utils.helper import Helper
```

```python
def get_frame_count(cached_processed_frames_path: Path) -> int:

    frames = Helper.load_python_object(cached_processed_frames_path)
    return len(frames)
```

```python
videos_df["frame_count"] = videos_df["cached_processed_frames_path"].apply(get_frame_count)
```

```python
videos_df.head()
```

```python
videos_df["duration"] = videos_df["frame_count"].apply(lambda x: x*3)
```

```python
videos_df
```

```python
def get_formatted_duration(seconds: int) -> str:
    minutes = seconds // 60
    seconds = seconds % 60
    return f"{minutes:02d}:{seconds:02d}"
```

```python
videos_df["formatted_duration"] = videos_df["duration"].apply(get_formatted_duration)
```

```python
frame_count_threshold = 30 * 60 // 3
```

```python
short_videos = videos_df[videos_df["frame_count"] <= frame_count_threshold]
```

```python
long_videos = videos_df[videos_df["frame_count"] > frame_count_threshold]
```

```python
def populate_video_id(df: pd.DataFrame) -> pd.DataFrame:
    df["video_id"] = df["video_name"].apply(Helper.get_video_id)
    return df
```

```python
def _save(df: pd.DataFrame, file_path: str) -> None:
    df.to_csv(file_path, sep="\t", index=False)
```

```python
output_dir = "./38"
output_dir = Path(output_dir)
output_dir.mkdir(parents=True, exist_ok=True)
short_output_path = output_dir / "short_videos.csv"
long_output_path = output_dir / "long_videos.csv"
```
